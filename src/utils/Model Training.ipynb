```python
from ultralytics import YOLO
from IPython.display import display, Image
```

```python
model = YOLO('yolov8l.pt')
```

```python
model.train(
    data='/content/REACHOBJ-3/data.yaml',  # Replace with the actual path to your data.yaml file
    epochs=10,
    batch=16,
    imgsz=640,
##    freeze=12
)
```

```python
metrics = model.val(
    data='/content/REACHOBJ-3/data.yaml',  # Path to your data.yaml file
    split='test'               # Specify the test split for evaluation
)
print(metrics)  # Outputs metrics such as mAP, precision, and recall
```

```python
# Evaluate model on the validation set
metrics = model.val(data='/content/REACHOBJ-3/data.yaml',  # Path to your data.yaml file
                    batch=16,
                    imgsz=640)
```
